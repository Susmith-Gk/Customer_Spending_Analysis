```python
import pandas as pd
import numpy as np

# Loading the dataset
df = pd.read_csv('spending_patterns_detailed.csv')
```

```python
df.head()
```

```python
# Checking data types and missing values
print(df.info())
print(df.isnull().sum())
```

```python
# Correcting data types
df['Customer ID']= df['Customer ID'].astype(str)
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])
```

```python
# Validating data (check for negative quantities or prices)
df = df[df['Quantity'] >= 0]
df = df[df['Price Per Unit'] >= 0]
```

```python
# Standardizing categorical data (e.g., consistent capitalization)
df['Category'] = df['Category'].str.title() # Capitalize first letter of each word.
df['Location'] = df['Location'].str.title()
df['Payment Method'] = df['Payment Method'].str.title()
```

```python
#Standardizing Tiles (removing spaces)
df.columns = df.columns.str.replace(' ', '_')
df.info()
```

```python
# Saving the cleaned data
df.to_csv('cleaned_transactions.csv')
```
